### Scatterplot with multiple semantics

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

##데이터 호출
diamonds = sns.load_dataset('diamonds')

##scatterplot
f,ax =plt.subplots(figsize=(6.5,6.5))
sns.despine(f, left=True, bottom=True) ##왼쪽과 아래 테두리 제거
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"] ##오름차순(IF가 최고등급)

sns.scatterplot(x='carat',y='price',hue='clarity',size='depth',
                palette='ch:r=-.2,d=.3_r', ##palette로 hue 색상 지정
                hue_order=clarity_ranking, 
                sizes=(1,8),linewidth=0,
                data=diamonds,ax=ax)
```

### TimeSeries plot with error band

```python
import seaborn as sns
sns.set_theme(style='darkgrid')

fmri = sns.load_dataset('fmri')

sns.lineplot(x='timepoint',y='signal',hue='region',style='event',data=fmri)
```

### displot  
#### Facetting histograms by subsets of data

```python
import seaborn as sns

sns.set_theme(style='darkgrid')

df = sns.load_dataset('penguins')
sns.displot(df,x='flipper_length_mm',col='species',row='sex',
            binwidth=3,height=3,facet_kws=dict(margin_titles=True))
```
